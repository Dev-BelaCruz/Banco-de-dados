```python
# Importando a biblioteca matplotlib para criar o gráfico
import matplotlib.pyplot as plt

# Definindo as variáveis
fatias = [8, 12, 10]  # Quantidade em cada categoria
atividades = ['Lazer', 'Estudo', 'Viagem']  # Rótulos das categorias
colunas = ['r', 'm', 'y']  # Cores para cada fatia

# Criando o gráfico de pizza
plt.pie(fatias,
        labels=atividades,
        colors=colunas,
        startangle=90,  # Iniciando o gráfico em um ângulo de 90°
        shadow=True,    # Adicionando sombra
        explode=(0, 0.1, 0)  # Destacando a segunda fatia (Estudo)
       )

# Exibindo o gráfico
plt.show()
```
