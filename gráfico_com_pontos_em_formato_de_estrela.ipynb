```python
# Importando a biblioteca matplotlib
import matplotlib.pyplot as plt

# Definindo as variáveis
x = [1, 2, 3, 4, 5, 6, 7, 8]  # Valores para o eixo x
y = [5, 2, 4, 5, 6, 8, 4, 8]  # Valores para o eixo y

# Criando o gráfico de dispersão (scatter plot) com pontos em formato de estrela
plt.scatter(x, y, label='Pontos', color='b', marker='*', s=100)

# Exibindo a legenda
plt.legend()

# Exibindo o gráfico gerado
plt.show()
```
