```python
# Importando a biblioteca matplotlib
import matplotlib.pyplot as plt

# Definindo as variáveis
x = [1, 3, 5]  # Valores para o eixo x
y = [1, 2, 5]  # Valores para o eixo y

# Criando o gráfico de linha
plt.plot(x, y)

# Atribuindo um título ao gráfico
plt.title('Exemplo utilizando Plot')

# Atribuindo rótulos aos eixos
plt.xlabel('Variável 1')  # Rótulo para o eixo x
plt.ylabel('Variável 2')  # Rótulo para o eixo y

# Atribuindo uma legenda ao gráfico
plt.plot(x, y, label='Uma legenda')

# Exibindo a legenda
plt.legend()

# Exibindo o gráfico gerado
plt.show()
```
